```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Load the data from CSV files
orders_df = pd.read_csv("C:/Users/Vaibhav/Desktop/inotask3/diminos_data_v2/orders.csv")

# Step 2: Preprocess the Data
orders_df['order_placed_at'] = pd.to_datetime(orders_df['order_placed_at'])


orders_df['hour'] = orders_df['order_placed_at'].dt.hour + orders_df['order_placed_at'].dt.minute / 60.0

# Step 3: Calculate the average number of orders per hour
lambda_value = orders_df.groupby('hour').size().mean()

# Generate x values (number of orders)
x = np.arange(0, 50)

# Step 4: Calculate probabilities for each value of x
probabilities = poisson.pmf(x, lambda_value)

# Step 5: Plot the Poisson distribution with increased resolution
plt.figure(figsize=(10, 6))
plt.plot(x, probabilities, marker='o', linestyle='-', color='skyblue', label=f'Lambda = {lambda_value:.2f}')
plt.title('Poisson Distribution of Orders Incoming Each Hour')
plt.xlabel('Number of Orders')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.xticks(np.arange(0, 51, 2))
plt.show()
```


In the Poisson distribution plot, the probability represents the likelihood of observing a certain number of orders per hour based on the Poisson distribution model, with the lambda parameter influencing the shape and characteristics of the distribution. Each point on the plot corresponds to a specific number of orders, and the height of the point (or the y-axis value) represents the probability of observing that number of orders.
